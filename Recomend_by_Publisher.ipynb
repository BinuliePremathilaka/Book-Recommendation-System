```python
def get_publisher_recommendations(publisher_name):
    # Filter books based on the given publisher
    publisher_books = books[books['publisher'] == publisher_name]

    # If no books found for the given publisher, return an empty list
    if publisher_books.empty:
        print(f'No data found for Publisher "{publisher_name}"')
        return []

    # Return the titles of recommended books
    return list(publisher_books['title'])

# Example usage
publisher_to_recommend = 'O\'Reilly Media'
publisher_recommendations = get_publisher_recommendations(publisher_to_recommend)

print(f'Recommended Books for Publisher "{publisher_to_recommend}":')
for book_title in publisher_recommendations:
    print(f'- {book_title}')
```
